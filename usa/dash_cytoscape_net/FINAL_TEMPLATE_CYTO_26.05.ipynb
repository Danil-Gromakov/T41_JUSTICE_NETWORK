```python
#Following & Followed nodes and edges
@app.callback([Output('output_json', 'children'),
              Output('cytoscape-net', 'stylesheet'),
              Output('img-prosecutor','children')],

              [Input('cytoscape-net', 'tapNode'),
               Input('cytoscape-net', 'selectedNodeData'),
               Input('dropdown-callbacks-label', 'value')])

def test(node,selectedNodeData,nodeLabel):
    if not node:
        new_styles = [{
            "selector": 'node',
            'style': {
                'opacity': 0.6,
                'content': nodeLabel
                #'background-image': 'data(url)'
                
            }
        }]
        return 'Node not chosen', default_stylesheet + new_styles + style_img_node ,html.H6('Node not chosen')
    

    if selectedNodeData == []:
        new_styles = [{
            "selector": 'node',
            'style': {
                'opacity': 0.6,
                'content': nodeLabel,
                #'background-image': 'data(url)'
            }
        }]
        return 'Node not chosen',default_stylesheet + new_styles + style_img_node ,html.H6('Node not chosen')
   
    
    
    if node:   
        stylesheet = [{
            "selector": 'node',
            'style': {
                'opacity': 0.4,
                'content': nodeLabel,
                'backgroundColor': '#bfbfbf'
                #'background-image': 'data(url)',
            }
        }, {
            'selector': 'edge',
            'style': {
                'opacity': 0.4,
                "curve-style": "bezier",
                'source-arrow-shape': 'triangle'
            }
        },
            {
        'selector': '.TRUMP',
        'style': {
            'width': 100,
            'height': 100
        }
    },{
        'selector': '.PROCONSERVATIVE',
        'style': {
            'width': 160,
            'height': 160
        }
    },
    {
        'selector': '.TRUMPALLY',
        'style': {
            'width': 80,
            'height': 80
        }
    }
        ]
        
        #tapNode change border
        stylesheet.append({
                    "selector": 'node[id = "{}"]'.format(node['data']['id']),
                    "style": {
                        'opacity': 1,
                        'shape': 'rectangle',
                        'border-width': 3,
                        'border-color': 'yellow',
                        'border-opacity':0.5,
                    }
                })

        #following
        #change color
        following_color = 'red'
        
        for edge in node['edgesData']:
             #node
            if edge['source'] == node['data']['id']:
                stylesheet.append({
                    "selector": 'node[id = "{}"]'.format(edge['target']),
                    "style": {
                        'background-color': following_color,
                        'opacity': 1,
                        'border-width': 3,
                        'border-color': following_color,
                        'border-opacity':1
                    }
                })
             #edge
          
                stylesheet.append({
                    "selector": 'edge[id= "{}"]'.format(edge['id']),
                    "style": {
                        "line-color": following_color,
                        'source-arrow-color':following_color,
                        
                    }
                })
        
        #follower
        #change color 
        follower_color = 'green'
        for edge in node['edgesData']:
             #node
            if edge['target'] == node['data']['id']:
                stylesheet.append({
                    "selector": 'node[id = "{}"]'.format(edge['source']),
                    "style": {
                        'background-color': follower_color,
                        'opacity': 1,
                        'border-width': 3,
                        'border-color': follower_color,
                        'border-opacity':1,
                    }
                })
           
            #edge
                stylesheet.append({
                    "selector": 'edge[id= "{}"]'.format(edge['id']),
                    "style": {
                        "line-color": follower_color,
                        'source-arrow-color': follower_color
                        
                    }
                })
                
        
        if int(node['data']['id']) < 500:
            img = html.Div([html.Img(src=node['data']['url'], style={'height':'30%', 'width':'30%'}),
                  html.Div(
                      [html.P(['Name: '+node['data']['label']]),
                       html.P(['Source: '+node['data']['source']]),
                       html.P(['Position_current: '+node['data']['position_current']]),
                       html.P(['Date_of_appointment: '+node['data']['date_of_appointment']])]

                  )])
        else:
            img = html.Div([html.Img(src=node['data']['url'], style={'height':'30%', 'width':'30%'}),
                  html.Div(
                      [html.P('Name: '+node['data']['label'])]

                  )])
            
                            
        
        return json.dumps(node['data'], indent=2), stylesheet + style_img_node ,img 
   # json.dumps(node['data'], indent=2), stylesheet + style_img_node ,img 
```
