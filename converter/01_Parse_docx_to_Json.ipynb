## Step 1: Parse docx to Json

```python
from docx import Document
```

```python
# preparing txt file with words and phrases which we don't want to see in docs
# below we will use it to clear data
text_no = open("text_no.txt","r").read()
```

```python

prosec = {}

# need to appoint every row some kind of id
r = -2
c = 0

#document = Document("prosecutors_2018.docx")
document = Document("sved_2019.docx")


for table in document.tables:
    for row in table.rows:
        r +=1
        name = []
        position = []
        income = []
        real_est_own = []
        real_est_use = []
        vehical = []
        for cell in row.cells:
            
            c+=1
            if c == 13:
                c = 1
            
            for paragraph in cell.paragraphs:
                
                if paragraph.text in text_no:
                    continue
                else:
                    
                    #check our output
                    print(r,c,paragraph.text)
                    
                    if c == 1:
                        name.append(paragraph.text)
                    elif c == 2:
                        position.append(paragraph.text)
                    elif c == 5:
                        real_est_own.append(paragraph.text)
                    elif c == 8:
                        real_est_use.append(paragraph.text)
                    elif c == 10:
                        vehical.append(paragraph.text)
                    elif c == 11:
                        income.append(paragraph.text)
                        
                prosec[r] = {"name":name,"position":position,
                             "income":income,"real_est_own":real_est_own,
                             "real_est_use":real_est_use,"vehical":vehical} 
            
            
            print("-----------------------------------------------------------------")
        #print(prosec)
```

```python
# To write at json

import json

with open ("2020/prosecutors_2019.json","w") as f:
    json.dump(prosec,f)
```
