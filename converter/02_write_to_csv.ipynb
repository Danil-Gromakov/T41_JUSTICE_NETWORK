## Step 2: Write from json to csv

```python
import json
import csv

with open('2020/prosecutors_2019.json') as json_file:
    data = json.load(json_file)
    
    #print(len(data[str(23)]["vehical"]))
    count_prosecutors = len(data)+1
```

```python
with open('2020/prosecutors_2019.csv', 'w', newline='') as csvfile:
    fieldnames = ['name',
                  'position',
                  #'sex',
                  'spouse',
                  'staf_income',
                  'spouse_income',
                  'minor_child',
                  'count_real_est',
                  'real_est_sq',
                  'vehical']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    


    for n in range(1,count_prosecutors):
        vehical = len(data[str(n)]["vehical"])
        
        #print(data[str(n)]["name"],data[str(n)]["position"])
        name = data[str(n)]["name"][0]
        
        position = data[str(n)]["position"]
        position = ' '.join([str(elem) for elem in position])
        
        
        if any("супруг" in s for s in data[str(n)]["name"]):
            spouse = data[str(n)]["name"][1]
            #if "супруг" == spouse:
             #   sex = "F"
            #else:
             #   sex = "M"
        else:
            spouse = ""
            #sex = ""
        
        if any("супруг" in s for s in data[str(n)]["name"]) and len(data[str(n)]["income"]) > 1 :
            
            inc = data[str(n)]["income"]
            
            staf_income = inc[0]
            spouse_income =inc[1]
            
            
        elif any("супруг" in s for s in data[str(n)]["name"]) and len(data[str(n)]["income"]) == 1 :
            inc = data[str(n)]["income"]
            staf_income = inc[0]
            spouse_income = "-1"
        else:
            inc = data[str(n)]["income"]
            staf_income = inc[0]
            
            spouse_income = ""
            
        if any("несоверш" in s for s in data[str(n)]["name"]):
            
            minor_child = "+"

        else:
            minor_child = ''
        
        all_real_est = data[str(n)]["real_est_own"] + data[str(n)]["real_est_use"]
        
        count_real_est = len(all_real_est)
        
        real_est_sq  = 0
        
        for n in all_real_est:
            n = n.replace(",",".").replace(" ","").replace(u'\xa0', u'')
            
            try:
                n = int(float(n))
            except:
                n = int(n)

            real_est_sq += n
        
        staf_income = staf_income.replace(" ","").replace(u'\xa0', u'')
        spouse_income = spouse_income.replace(" ","").replace(u'\xa0', u'')
        
        print(name)

        writer.writerow({'name': name,
                         'position': position,
                         #'sex':sex,
                         'spouse':spouse,
                         'staf_income':staf_income,
                         'spouse_income':spouse_income,
                         'minor_child':minor_child,
                         'count_real_est':count_real_est,
                         'real_est_sq':real_est_sq,
                         'vehical':vehical})
```
